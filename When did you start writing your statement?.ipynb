```python
%matplotlib inline
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(font_scale=2)
```

```python
responses = {
    "Sebastian": ("October", 10),
    "Roma": ("August", 8),  # Late August, early September.
    "Nicholas": ("August", 8),  # Late August, early September.
    "Nelson": ("September", 9),
    "Michi": ("September", 9),
    "Lucy": ("October", 10),
    "Kevin": ("September", 9),
    "Kalpesh": ("November", 11),
    "Eric": ("October", 10),
    "Akari": ("September", 9),
    "Aishwarya": ("September", 9)
}
start_date_counts = Counter(responses.values())
```

```python
plt.figure(figsize=(10, 5))
months_with_numbers, counts = zip(*sorted(start_date_counts.items(), key=lambda x: x[0][1]))
months, _ = zip(*months_with_numbers)
plt.bar(months, counts, align='center')
plt.title("When did you start writing your statement?")
plt.xlabel("Month Started")
plt.ylabel("Number of respondents")

plt.savefig("statement_start_date.png", bbox_inches="tight", pad_inches=0)
```
